```python
def aapt_perm(apks_path, perm_path):
    count = 0

    for (path, dir, files) in os.walk(apks_path):
        for filename in files:
            ext = os.path.splitext(filename)[-1]
            if ext == ".apk":
                fullfile_path = path+"/"+filename
                save_path = perm_path + filename[:-4] +".txt"
                
                if os.path.isfile(save_path):
                    print(count, "File exist!" + filename)
                    
                else:
                    terminal = f"aapt dump permissions {fullfile_path} > {save_path}"
                    new = subprocess.run(terminal, shell = True)
                    print(count, "done", filename)
                count += 1
```
